```python
import numpy as np
```

### create a numpy array of size 100 equal elements in between -50 to 50

```python
x=np.linspace(-50,50,100)
x
```

### create another array of same size with sin value of x

```python
y=np.sin(x)
y
```

### now you can plot a sin graph using "MatPlotLib" library in python

```python
# we only import pyplot function

import matplotlib.pyplot as plt
```

```python
%matplotlib inline          
# to hold & print the graph on screen
```

#### sin wave graph

```python
# inside pyplot there is a function - plot()

plt.plot(x,y)        # pass the array of data which is apper on graph
```

```python
plt.plot(x)       
```

#### cos wave graph

```python
a = np.cos(x)
a
```

```python
plt.plot(a)
```

#### Quadratic Eq. graph (parabolic)

```python
z = x*x + 2*x + 6
z
```

```python
plt.plot(z)
```

#### Cubic Eq. graph (cubic curve)

```python
b = x*x*x + 2*x*x + 6*x + 12
b
```

```python
plt.plot(b)
```
